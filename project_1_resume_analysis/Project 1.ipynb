3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
# создаем функцию преобразования строки с опытом работы в число месяцев опыта работы
def experience_months(text):
    # если тип данных не строка (встречаются другие типы данных)
    if type(text) is not str:
        return np.NaN
    # обязательно должен быть строковый тип данных
    else:
        # если опыт не указан - возвращаем nan
        if text == 'Не указано':
            return np.NaN
        # во всех остальных случаях
        else:
            # строку преобразуем в список и оставляем первые 6 элементов, вся инфа должна быть в [0:5]
            exp_list = text.split(' ')[:6]
            # задаем исходное значение опыта в месяцах
            exp_months = 0
            
            # ищем опыт в годах
            # слово год может быть записано следующим ообразом
            years = ('лет', 'года', 'год')
            # создаем цикл поиска элементов кортежа выше в исходном списке
            for i in years:
                if i in exp_list:
                    # определяем позицию элемента в списке
                    pos = exp_list.index(i)
                    # значение опыта в годах всегда на одну позицию меньше и приводим к типу float
                    year = float(exp_list[pos - 1])
                    # значение опыта увеличиваем на опыт в годах * 12                    
                    exp_months += year * 12
                    
            # ищем опыт в месяцах
            # слово месяц может быть записано следующим ообразом
            months = ('месяцев', 'месяц', 'месяца')  
            # создаем цикл поиска элементов кортежа выше в исходном списке
            for i in months:
                if i in exp_list[:6]:
                    # определяем позицию элемента в списке
                    pos = exp_list.index(i)
                    # значение опыта в месяцах всегда на одну позицию меньше и приводим к типу float
                    month = float(exp_list[pos - 1])
                    # значение опыта увеличиваем на опыт в месяцах  
                    exp_months += month
                    
            # результат работы функции - вернуть опыт в месяцах
            return exp_months


# в df создаем новый признак опыта работы в месяцах
df['Опыт работы (месяц)'] = df['Опыт работы'].apply(experience_months)

# удаляем столбец с опытом работы
df = df.drop(['Опыт работы'], axis=1)
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
# создаем функцию выделения признака города, переезда и командировок из строки
def city_move_trips(text):
    main_list = text.split(',')
    # разбираемся с городом
    # города милионники
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 
                      'Казань', 'Челябинск', 'Омск', 'Самара', 
                      'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]
    # город это первый элемент списка
    city = main_list[0].rstrip()
    if city in ('Москва', 'Санкт-Петербург'):
        city = city
    elif city in million_cities:
        city = 'город-миллионник'
    else:
        city = 'другие'
    
    # поиск готовности к переезду и командировкам, по умолчанию не готовы
    # ready_to_move = False
    ready_to_trip = False
    
    for elem in main_list:
        # по переезду
        if 'перее' in elem:
            # проверяем на соответсвие первый символ
            # если начинается с Н - не готов, г - готов, х - хочет
            if ('не готов к переезду' in elem)  or ('не готова к переезду' in elem):
                ready_to_move = False
            else:
                ready_to_move = True
        
        # по командировкам
        if 'команд' in elem:
            # проверяем на соответсвие первый символ
            # если начинается с Н - не готов, г - готов, х - хочет
            if ('не готов к команд' in elem)  or ('не готова к команд' in elem):
                ready_to_trip = False
            else:
                ready_to_trip = True
    
    
        
    
    return city, ready_to_move, ready_to_trip
    
# создаем новые признаки (применяем функцию к признаку 'Город, переезд, командировки', а потом лямбдой вычленяем нужный элемент из кортежа)
df['Город'] = df['Город, переезд, командировки'].apply(city_move_trips).apply(lambda x: x[0])
df['Готовность к переезду'] = df['Город, переезд, командировки'].apply(city_move_trips).apply(lambda x: x[1])
df['Готовность к командировкам'] = df['Город, переезд, командировки'].apply(city_move_trips).apply(lambda x: x[2])

# для задания 3.4
# процент соискателей из Питера
# display(df['Город'].value_counts(normalize=True)['Санкт-Петербург'])

# => 0.11033881637761488

# процент готовых к переезду и командировкам
# mask1 = (df['Готовность к переезду'] == True)
# mask2 = (df['Готовность к командировкам'] == True)
# display(round(df[mask1 & mask2].shape[0] / df.shape[0] * 100))

# => 32%

# удаляем более не нужный признак
df = df.drop(['Город, переезд, командировки'], axis=1)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# создаем функцию преобразования занятости
def employment(text):
    # полная занятость
    full_empl = False
    if 'полная' in text:
        full_empl = True
    
    # частичная занятость
    part_empl = False
    if 'частич' in text:
        part_empl = True
        
    # проектная работа
    proj_work = False
    if 'проект' in text:
        proj_work = True
        
    # стажировка
    intership = False
    if 'стажир' in text:
        intership = True
        
    # волонтерство
    volunt = False
    if 'волонт' in text:
        volunt = True
        
    return  full_empl,\
            part_empl,\
            proj_work,\
            intership,\
            volunt

# добавляем новые признаки в датафрейм
df['полная_занятость'] = df['Занятость'].apply(employment).apply(lambda x: x[0])
df['частичная_занятость'] = df['Занятость'].apply(employment).apply(lambda x: x[1])
df['проектная_работа'] = df['Занятость'].apply(employment).apply(lambda x: x[2])
df['стажировка'] = df['Занятость'].apply(employment).apply(lambda x: x[3])
df['волонтерство'] = df['Занятость'].apply(employment).apply(lambda x: x[4])

# ответ на задание 3.5
mask1 = df['проектная_работа']
mask2 = df['волонтерство']
df[mask1 & mask2].shape
# (436, 20)
```

```python
# функция преобразования графика работы
def schedule(text):
    # гибкий график
    flex_sched = False
    if 'гибкий' in text:
        flex_sched = True
    
    # полный день
    full_day_sched = False
    if 'полный' in text:
        full_day_sched = True
        
    # сменный график
    shift_work_sched = False
    if 'сменный' in text:
        shift_work_sched = True
        
    # вахтовый метод
    shift_method = False
    if 'вахтовый' in text:
        shift_method = True
        
    # удаленная работа
    distant_work = False
    if 'удаленная' in text:
        distant_work = True
        
    return  flex_sched,\
            full_day_sched,\
            shift_work_sched,\
            shift_method,\
            distant_work
            
# добавляем новые признаки в датафрейм

df['гибкий_график'] = df['График'].apply(schedule).apply(lambda x: x[0])
df['полный_день'] = df['График'].apply(schedule).apply(lambda x: x[1])
df['сменный_график'] = df['График'].apply(schedule).apply(lambda x: x[2])
df['вахтовый_метод'] = df['График'].apply(schedule).apply(lambda x: x[3])
df['удаленная_работа'] = df['График'].apply(schedule).apply(lambda x: x[4])

# ответ на задание 3.5
# mask1 = df['гибкий_график']
# mask2 = df['вахтовый_метод']
# df[mask1 & mask2].shape
# (2311, 25)

# удаляем из датафрейма признаки занятость и график
df = df.drop(['Занятость', 'График'], axis=1)
```

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
# подгружаем датафрейм с курсом валют
df_curr = pd.read_csv('data/ExchangeRates.csv')
```

```python
# переводим в формат даты
df['Обновление резюме'] = pd.to_datetime(df['Обновление резюме'], dayfirst=True)
df_curr['date'] = pd.to_datetime(df_curr['date'], dayfirst=True)
# приводим даты к одинаковому формату (т.к в исходной таблице еще время)
df['Обновление резюме'] = df['Обновление резюме'].dt.date
df['Обновление резюме'] = pd.to_datetime(df['Обновление резюме'])

# оставляем из таблицы валют только нужные признаки
df_curr = df_curr[['currency', 'date', 'proportion', 'close']]

# функция преобразования желаемой зп в рубли
def salary_to_rub(text):
    '''
    на вход принимает текст из строки,
    возвращает сумму и наименование валюты, согласно изо-коду
    '''
    # получаем из текста сумму и валюту
    salary, currency = int(text.split()[0]), text.split()[1]
    
    # словарь изо-кодов исходя из конкретно тех кодов, которые нам нужны для решения частного случая
    curr_iso_code = {
    'руб.':'RUB',
    'бел.руб.':'BYN',
    'грн.':'UAH',
    'сум':'UZS',
    }
    
    if currency in curr_iso_code:
        currency = curr_iso_code[currency]
    
    return salary, currency


# самым очевидным вариантом кажется создать датафрейм, который будет результатом объединения основного ДФ с ДФ валют.
# предварительно в основной датафрейм применим функцию к признаку ЗП и получим два новых столбца с суммой и с наименованием валюты
# объединяем методом left
# из полученного датафрейма оставляем только необходимые для расчета признаки

# добавляем признаки в исходный фд
df['зарплата_сумма'] = df['ЗП'].apply(salary_to_rub).apply(lambda x: x[0])
df['зарплата_валюта'] = df['ЗП'].apply(salary_to_rub).apply(lambda x: x[1])

# создали таблицу объединив необходимые признаки из первой таблицы с соответсвующими признаками из второй
# оставляем только необходимые признаки
merged = df.merge(
    df_curr,
    left_on=['зарплата_валюта', 'Обновление резюме'],
    right_on=['currency', 'date'],
    how='left'
)[['зарплата_сумма', 'зарплата_валюта', 'proportion', 'close']]


# там, где ЗП была указана в рублях, по некоторым признакам в результате объединения получились NaN
# для заполнения пустых значений создаем словарь, в котором указываем какие признаки на какую величину заменять
values = {'proportion': 1,
          'close': 1}

merged = merged.fillna(value=values)
# и добавляем итоговый признак с зп в рублях
merged['salary'] = merged['зарплата_сумма'] / merged['proportion'] * merged['close']

# данные по зарплате в рублях перекидываем в исходную таблицу
df['зп_руб'] = merged['salary']

# для ответа на задание 3.6
display(df['зп_руб'].median())

# удаляем лишние столбцы
df = df.drop(['ЗП', 'зарплата_сумма', 'зарплата_валюта'], axis=1)
```

Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
# np.log(df['Возраст']).skew()
# 0.4523422276053838
# логарифмическое распределение ассиметрично в правую сторону

# создаем функцию поиска выбросов методом z-отклонений
# с послаблением сигмы на 1 в правую сторону
def outliers_z_score(data, feature, log_scale=False):
    if log_scale:
        x = np.log(data[feature])
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - 3 * sigma
    upper_bound = mu + 4 * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x > lower_bound) & (x < upper_bound)]
    return outliers, cleaned, mu, lower_bound, upper_bound

outliers, cleaned, mu, lower_bound, upper_bound = outliers_z_score(df, 'Возраст', True)
histplot = sns.histplot(
    # data=df,
    x=np.log(df['Возраст']),
    bins=50
)
histplot.axvline(mu, color='k', lw=2)
histplot.axvline(lower_bound, color='k', lw=2, ls='--')
histplot.axvline(upper_bound, color='red', lw=2, ls='--');
histplot.set_title('Распределение возраста соискателей в логарифмическом масштабе,\n с границами, согласно методу z-отклонений');

# результирующий график, который показан сейчас - построен по данным уже с удаленными выбросами
```
